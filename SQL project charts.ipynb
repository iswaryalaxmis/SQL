```python
df8 = pd.read_csv("/content/drive/MyDrive/GREAT LEARNING/AIML/SQL/Project/SQL proj/Tables/8.csv")
df8
```

```python
# Define colors for the lines
revenue_color = 'b'  # Blue
order_color = 'g'  # Green

# Create the primary y-axis (revenue)
fig, ax1 = plt.subplots()
ax1.plot(df8['quarter_number'], df8['Quarterly_Revenue'], color=revenue_color, label='Revenue')
ax1.set_xlabel('Quarter')
ax1.set_ylabel('Revenue', color=revenue_color)
ax1.tick_params(axis='y', labelcolor=revenue_color)

# Create the secondary y-axis (orders)
ax2 = ax1.twinx()  # Share the x-axis
ax2.plot(df8['quarter_number'], df8['Count_orders'], color=order_color, marker='o', linestyle='--', label='Orders')
ax2.set_ylabel('Count_orders', color=order_color)
ax2.tick_params(axis='y', labelcolor=order_color)

# Customize the plot (optional)
plt.title('Trend of Revenue and Orders by Quarter')
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
plt.legend(lines1 + lines2, labels1 + labels2, loc='upper right')
plt.grid(True)
plt.show()
```
